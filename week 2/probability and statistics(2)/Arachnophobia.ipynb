```python
## packages
import numpy as np 
import sklearn
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## set your data
x = np.array([3, 4.3, 9.2, 7.6, 13, 6.4, 9.9, 3.3, 5.6, 12, 10.4, 13.9, 5.9, 4.7,
9.5, 11.3, 15, 11.9, 10.7, 6.5, 5.3, 11.6, 14.8, 8.9])

y = np.array([15, 20, 30, 25, 35, 18, 40, 10, 11, 28, 26, 36, 27, 17, 34, 41, 50,
43, 37, 21, 14, 38, 45, 24])
```

```python
## correlation coefficient between x and y
np.corrcoef(x, y)
```

```python
## reshape x
x = x.reshape((-1, 1))
```

```python
linear_model = LinearRegression().fit(x, y)
```

```python
# intercept
linear_model.intercept_
```

```python
# coefficient
linear_model.coef_
```

```python
# predictions
y_pred = linear_model.predict(x)
```

```python
# residuals
e = y - y_pred

print(e)
```

```python
# plot the unfitted data
plt.scatter(x, y)
plt.title('Unfitted plot of x and y')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# plot
#sns.scatterplot(x= x, y= y)
```

```python
# fit the data
plt.plot(x, y_pred, color= 'r')
plt.scatter(x, y)
plt.title('fitted plot of x and y')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# score
R_score = linear_model.score(x, y)

print(R_score)
```
